```python
from datetime import datetime

Time_difference = []
date_format_str = '%Y-%m-%d %H:%M:%S'
data_10 = data_10.dropna()
data_10 = data_10.reset_index(drop=True)

for i in range(len(data_10['DATE_REPORTED'])) :
    date_1 = str(data_10['DATE_REPORTED'][i])
    date_2 = str(data_10['DATE_OCCURED'][i])
    start = datetime.strptime(date_2, date_format_str)
    end =   datetime.strptime(date_1, date_format_str)
    diff = end - start
    Time_difference.append(diff.total_seconds() / 3600)
    
data_10new = data_10
data_10new['Time_difference'] = Time_difference
data_10new
```
